```python
df2 = pd.read_csv('/Users/thananpornsethjinda/Desktop/pulsar-pred/data/pulsar_data_test.csv')
 
```

```python
first_row = pd.DataFrame([df.columns.to_list()], columns=df.columns)

print(first_row)

actual_df = pd.concat([first_row, df]).reset_index(drop=True)

actual_df.columns = df2.columns
```

```python
actual_df
```

```python
# making train test split (stratified)

from sklearn.model_selection import train_test_split

print(actual_df.columns[0:-1])

X = actual_df[actual_df.columns[0:-1]]

def to_float(x):
    return float(x)

X = X.applymap(to_float)

y = actual_df[actual_df.columns[-1]]

def to_int(x):
    return int(x)

y = y.apply(to_int)

y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()
X_train_processed = scale.fit_transform(X_train)
X_test_processed = scale.transform(X_test)

```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(X_train_processed, np.array(y_train))

lr.coef_
```

```python
y_pred = lr.predict(X_test_processed)
```

```python
from sklearn.metrics import roc_auc_score, confusion_matrix

roc_auc_score(y_true=y_test, y_score=y_pred)

confusion_matrix(y_true=y_test, y_pred=y_pred, normalize='true')

```

```python
def sigmoid(y): 
    return 1/(1+np.e**(-y))

def f(t1, t2, t3, t4, t5, t6, t7, t8):

    loss_func = []

    z = np.array([t1, t2, t3, t4, t5, t6, t7, t8])

    for index in range(len(X_train_processed)): 

        h = sigmoid(np.dot(z, X_train_processed[index]))

        h = np.clip(h, 1e-15, 1 - 1e-15)

        y = np.array(y_train)[index]

        loss_func.append(-np.log(h**y*(1-h)**(1-y)))

    return np.sum(loss_func)
```
